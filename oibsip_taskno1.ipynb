```python
import pandas as pd
import folium
from folium.plugins import HeatMap
```

```python
df=exel=pd.read_csv("C:\\Users\\Sambh\\Downloads\\archive (1)\\Unemployment_Rate_upto_11_2020.csv")
df
```

```python
lats_longs_weight1= list(map(list, zip(df["lon"], df["lat"], df["UnempRate"])))
```

```python
map_obj1 = folium.Map(location =[20.5937,78.9629], zoom_start=5)
HeatMap(lats_longs_weight1).add_to(map_obj1)
print("Unemployment Rate:")
map_obj1
```

```python
lats_longs_weight2= list(map(list, zip(df["lon"], df["lat"], df["EstimatedEmployed"])))
```

```python
map_obj2 = folium.Map(location =[20.5937,78.9629], zoom_start=5)
HeatMap(lats_longs_weight2).add_to(map_obj2)
print("People Empolyed")
map_obj2
```

```python
lats_longs_weight3= list(map(list, zip(df["lon"], df["lat"], df["EstimatedLabourParticipationRate"])))
```
