```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
dfT3=pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/CarPrice.csv")
dfT3
```

```python
dfT3.isnull().sum()
```

```python
dfT3.describe()
```

```python
dfT3.CarName.unique()
```

```python
dfT3.corr()
```

```python
plt.figure(figsize=(20, 15))
correlations = dfT3.corr()
sns.heatmap(correlations, cmap="coolwarm", annot=True)
plt.show()
```

```python
predict = "price"
dfT3 = dfT3[["symboling", "wheelbase", "carlength", "carwidth", "carheight", "curbweight", "enginesize", "boreratio", "stroke", "compressionratio", "horsepower", "peakrpm", "citympg", "highwaympg", "price"]]
x = np.array(dfT3.drop([predict], 1))
y = np.array(dfT3[predict])
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
```

```python
model = DecisionTreeRegressor()
model.fit(xtrain, ytrain)
predictions = model.predict(xtest)
```

```python
model.score(xtest, predictions)
```
