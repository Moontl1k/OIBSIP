```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dfT5=pd.read_csv("C:\\Users\\Sambh\\Downloads\\archive (2)\\Advertising.csv")
dfT5
```

```python
dfT5=dfT5[["TV", "Radio", "Newspaper", "Sales"]]
dfT5
```

```python
dfT5.isnull().sum()
```

```python
dfT5.describe()
```

```python
dfT5.corr()
```

```python
plt.figure(figsize=(8, 4))
correlationsT5 = dfT5.corr()
sns.heatmap(correlationsT5, cmap="coolwarm", annot=True)
plt.show()
```

```python
predictT5 = "Sales"
xT5 = np.array(dfT5.drop([predictT5], 1))
yT5 = np.array(dfT5[predictT5])
```

```python
xtrainT5, xtestT5, ytrainT5, ytestT5 = train_test_split(xT5, yT5, test_size=0.2)
```

```python
modelT5 = LinearRegression()
modelT5.fit(xtrainT5, ytrainT5)
ypredect = modelT5.predict(xtestT5)
```

```python
plt.scatter(xtrainT5[:,0], ytrainT5,color='r') 
plt.scatter(xtrainT5[:,1], ytrainT5,color='g') 
plt.scatter(xtrainT5[:,2], ytrainT5,color='b') 
plt.plot(xtestT5, ypredect,color='k') 
plt.legend(["TV" , "Radio", "Newspaper","Prediction"])
plt.show()
```

```python
modelT5.score(xtestT5, ypredect)
```
